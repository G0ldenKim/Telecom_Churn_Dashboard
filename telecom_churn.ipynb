```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("telecom_churn.csv")
df.head()
```

```python
df = df.dropna()
df["Churn"] = df["Churn"].map({"Yes":1, "No":0})
```

```python
df["Churn"].value_counts()
```

## Data Cleaning
- Removed missing values
- Converted churn column into numeric format

## Data Analysis
The following visualizations explore factors affecting customer churn.

```python
sns.countplot(x="Contract", hue="Churn", data=df)
plt.show()
```

## Key Insights

- Customers on month-to-month contracts have higher churn rates  
- Higher monthly charges are associated with increased churn  
- Customers with shorter tenure are more likely to leave  

```python
sns.boxplot(x="Churn", y="MonthlyCharges", data=df)
plt.show()
```

```python
sns.boxplot(x="Churn", y="tenure", data=df)
plt.show()
```

## Recommendations

- Encourage long-term contracts through promotions  
- Offer discounts for high-paying customers  
- Focus retention strategies on new customers
